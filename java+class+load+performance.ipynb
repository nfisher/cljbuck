I've started on something of a rather large experiment to explore how I can improve Clojure start-up time. I was curious what impact compression level has on start-up latency for JVM applications. Four Zip compression levels were tested 0, 1, 6, and 9. The uberjar was unpacked and then repacked using the linux command line zip tool as follows:

```
zip -r -0 ../main_c0.jar *
zip -r -1 ../main_c1.jar *
zip -r -6 ../main_c6.jar *
zip -r -9 ../main_c9.jar *
```

All applications, screen-locks, system sleep, and anti-virus were closed to minimise interference. A test was devised which executes each test jar file 100 times per run. Each run has a sleep of 30s to settle the system and a write to disk of an 8MB file to flush disk caches following every 10 executions. See the test runner shell script for exact details. The test was left to run and the results from the first run were discarded as a "warm-up" cycle. In the future I would probably reduce the number of runs in the warm-up to 10 or 20.

## Hypothesis

- H0 - compression level will not have an impact on application start-up time.
- Ha - compression level will have an impact on application start-up time.
- p-level - 0.1.

## Results

The results are below in ms for the execution time.

### Run 2 Results

| (ms)   | C0  | C1  | C6  | C9  |
|--------|-----|-----|-----|-----|
| 25PCTL | 570 | 616 | 613 | 613 |
| 75PCTL | 600 | 655 | 643 | 643 |
| Max    | 650 | 730 | 702 | 700 |

As can be seen above compression level 0 does have an impact which is statistically significant and appears to shave 50ms from load time.

## Next Steps

Compressing with level 0 has an obvious tradeoff for the distribution of artefactos over the internet. I'd be curious to see what impact a faster algorithm such as Snappy or LZ4 has on class load performance.

For the next test I'll see what happens when the java byte code is loaded into a byte array and a custom class loader is used.

	while [ $B -lt 10 ]; do
		# try to flush the disk cache

````python
%matplotlib inline

import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

rounds = 2 # ignore the first round
levels = ['0', '1', '6', '9']
runs = []
for i in range(len(levels)*2):
    col = i % len(levels)
    run = 2 + math.floor(i/len(levels))
    level = levels[col]
    
    filename = 'main_c{}.jar-results-r{}.txt'.format(level, run)
    df = pd.read_csv(filename, header=None)
    df.columns = [level]
    if col == 0:
        runs.append(df)
    else:
        runs[run-2] = pd.concat([runs[run-2], df], axis=1, join='inner')
````

````python
second = runs[0]
third = runs[1]

result = stats.ttest_ind(second['0'].as_matrix(), second['1'].as_matrix(), equal_var = False)
print("significant:", result.pvalue < 0.01)
result = stats.ttest_ind(third['0'].as_matrix(), third['1'].as_matrix(), equal_var = False)
print("significant:", result.pvalue < 0.01)
````
